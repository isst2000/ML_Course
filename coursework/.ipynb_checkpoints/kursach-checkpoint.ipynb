```python
overshoots_big = []
overshoots_low = []

for index in range(data.shape[0]):
    if data[' shares'][index]>=50000: 
        overshoots_big.append(index)
    if data[' shares'][index]>10000 and data[' shares'][index]<50000: 
        overshoots_low.append(index)
print("Репостов больше 50 000: {}".format(len(overshoots_big)))
print("Репостов >50 тыс по отношению ко всем {}".format(len(overshoots_big)/len(data[' shares'])))
print("Репостов больше 10 000 но меньше 50 000: {}".format(len(overshoots_low)))
print("Репостов >10 тыс по отношению ко всем {}".format(len(overshoots_low)/len(data[' shares'])))
```

```python
def errors_methrics(target):
    print("Средняя абсолютная ошибка {}".format(mean_absolute_error(y_test, target)))
    print("Cредняя квадратичная ошибка {}".format(mean_squared_error(y_test, target)))
    print("Медианная абсолютная ошибка {}".format(median_absolute_error(y_test, target)))
```

```python
def compare_methrics(best,baseline):
    deltha_mae=mean_absolute_error(y_test, baseline)-mean_absolute_error(y_test, best)
    deltha_mse=mean_squared_error(y_test, baseline)-mean_squared_error(y_test, best)
    deltha_medae=median_absolute_error(y_test, baseline)-median_absolute_error(y_test, best)
    print("Средняя абсолютная ошибка меньше на {}".format(deltha_mae))
    print("Cредняя квадратичная ошибка меньше на {}".format(deltha_mse))
    print("Медианная абсолютная ошибка меньше на {}".format(deltha_medae,6))
```

```python
bestKNN = KNeighborsRegressor(n_neighbors=45)
bestKNN.fit(X_train, y_train)
bestTargetKNN = bestKNN.predict(X_test)

errors_methrics(bestTargetKNN)
```

```python
MAE=[]
MSE=[]
MEDAE=[]

for target,model_name in zip([bestTargetKNN, targetLR, bestTargetLasso, targetBestRidge, bestTargetDtr, bestRandForestTarget,bestGB_target],["K соседей","Линейная регрессия","Лассо","Гребневая регрессия","Дерево решений","Случайный лес","Градиентный бустинг"]):
    MAE.append([model_name, mean_absolute_error(y_test, target)])
    MSE.append([model_name, mean_squared_error(y_test, target)])
    MEDAE.append([model_name, median_absolute_error(y_test, target)])
```

```python
print("Качество моделей по метрике MAE")
sorted(MAE,key=lambda x: x[1])
```

```python
print("Качество моделей по метрике MSE")
sorted(MSE,key=lambda x: x[1])
```

```python
print("Качество моделей по метрике MEDAE")
sorted(MEDAE,key=lambda x: x[1])
```
